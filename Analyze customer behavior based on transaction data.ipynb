```python
data1=data[['QTY','VALUE']]
f=data1.sample(n=50000,replace=True)
f
```

```python
plt.boxplot(f,vert=False)
plt.title('detecting outliers across sample')
plt.show()
```

```python
plt.boxplot(data1,vert=False)
plt.title('detecting outliers across all the data set')
plt.show()
```

```python
lower_limit=data['VALUE'].mean()-3*data['VALUE'].std()
upper_limit=data['VALUE'].mean()+3*data['VALUE'].std()
new=data[(data['VALUE']>lower_limit)&(data['VALUE']<upper_limit)]
data1=new[['QTY','VALUE']]
f=data1.sample(n=50000,replace=True)
new
```

```python
plt.boxplot(data1,vert=False)
plt.title('detecting outliers across all the dataset after filtration')
plt.show()
```
